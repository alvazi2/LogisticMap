# Logistic Map - Numerical

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# define the logistic map function
LMap = lambda x, L: L*x*(1-x)
```

```python
# Plot various iterations starting from x = Seed
Seed = 0.5
LowL, HighL = (2.8, 4.0)
IterationFrom, IterationCount = (5000, 30)
IterationTo = IterationFrom + IterationCount
NumberOfValues = 50000
L_Values = np.linspace(LowL, HighL, NumberOfValues)
x_Values = np.linspace(Seed, Seed, NumberOfValues)
# initial iterations
for Iteration in range (IterationFrom):
    x_Values = LMap(x_Values, L_Values)
plt.figure(figsize=(20,15))
PointSize = 0.001
# iterations for plotting
for Iteration in range (IterationFrom, IterationTo): 
    p = plt.scatter(L_Values, x_Values, s=PointSize, color='black')
    x_Values = LMap(x_Values, L_Values)
```
